```python
import requests
from bs4 import BeautifulSoup

def extract_info(url):
    """Fetches and parses a webpage to extract information.

    Args:
        url: The URL of the webpage to crawl.

    Returns:
        A dictionary containing extracted information (if any), or None if unsuccessful.
    """
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise exception for unsuccessful requests
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {e}")
        return None

    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract desired information here (replace with your logic)
    # This example extracts the first 5 paragraphs
    contents = []
    for content in soup.find_all('p'):
        contents.append(content.text.strip())
        if len(contents) >= 5:
            break

    # Return extracted information
    return {"contents": contents, "soup": soup}

def print_extracted_data(data):
    """Prints the extracted data in a formatted way.

    Args:
        data: The dictionary containing extracted information.
    """
    if not data:
        print("No data extracted.")
        return

    print("Extracted Contents:")
    for content in data["contents"]:
        print(f" - {content}")

# Get URL from user
target_url = input("Enter website URL: ")

# Start crawling
extracted_data = extract_info(target_url)

print("Crawling complete. Extracted information:")
# Process and print extracted data (modify based on your needs)
print_extracted_data(extracted_data)
```
